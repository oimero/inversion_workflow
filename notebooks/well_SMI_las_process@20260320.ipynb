## 批量处理 SMI LAS 的 ~Curve 段

遍历 `data/vertical_well_las_target` 下全部 `.las/.Las` 文件，调用 `replace_smi_las_curve_section_in_file` 做原地替换。


```python
import sys
from pathlib import Path

repo_root = Path.cwd().resolve().parent
src_dir = repo_root / "src"
if str(src_dir) not in sys.path:
    sys.path.insert(0, str(src_dir))

from cup.utils.replace import replace_smi_las_curve_section_in_file

target_dir = repo_root / "data" / "vertical_well_las_target"
candidates = [*target_dir.glob("*.las"), *target_dir.glob("*.Las"), *target_dir.glob("*.LAS")]

# Windows 下 glob 可能重复命中同一文件，这里按绝对路径去重。
seen = set()
las_files = []
for p in candidates:
    key = str(p.resolve()).lower()
    if key not in seen:
        seen.add(key)
        las_files.append(p)
las_files = sorted(las_files)

print(f"repo_root: {repo_root}")
print(f"target_dir: {target_dir}")
print(f"发现 LAS 文件数量(去重后): {len(las_files)}")
for p in las_files:
    print(" -", p.name)
```

```python
processed = []
failed = []

for las_path in las_files:
    try:
        replace_smi_las_curve_section_in_file(las_path)
        processed.append(las_path)
    except Exception as e:
        failed.append((las_path, str(e)))

print(f"成功替换: {len(processed)}")
for p in processed:
    print(" [OK]", p.name)

print(f"失败: {len(failed)}")
for p, msg in failed:
    print(" [FAILED]", p.name, "->", msg)
```
